# Plotting data

```octave
t = [0:0.01:0.98]
```

```octave
y1 = sin(2 * pi * 4 * t);
plot(t, y1)
```

```octave
y2 = cos(2 * pi * 4 * t);
plot(t, y2)
```

```octave
# together
plot(t, y1);
hold on;
plot(t, y2);
xlabel("time");
ylabel("value");
legend("sin", "cos");
title("demo curve");
```

```octave
# together
plot(t, y1);
hold on;
plot(t, y2);
xlabel("time");
ylabel("value");
legend("sin", "cos");
title("demo curve");
print -dpng 'myPlot.png';   # print to file
close;                       # close the figure
```

```octave
figure(1); plot(t, y1);
figure(2); plot(t, y2);
```

```octave
subplot(1, 2, 1); % divide plot a 1*2 grid, access first element
plot(t, y1);
subplot(1, 2, 2);
plot(t, y2);
axis([0.5 1 -1 1]);         # 设置第二幅图的坐标范围
```
